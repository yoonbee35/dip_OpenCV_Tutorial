```python
#마우스로 점을 입력받아 ProspectiveTransform (더블클릭으로 원 생성)
import cv2
import numpy as np

points = []

# 왼쪽 상단, 오른쪽 상단, 왼쪽 하단, 오른쪽 하단 순으로 클릭하시오.

# mouse callback function
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global points
        cv2.circle(img,(x,y),10,(0,0,255),-1)
        print(x,y)
        points.append([x,y])
        
        
# Create a black image, a window and bind the function to window
img = cv2.imread('arashi.png')
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)

while(1):
    cv2.imshow('image',img)
    if cv2.waitKey(20) & 0xFF == 27: # enter ESC
        break
    if len(points) == 4:
        pts1 = np.float32(points)
        pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
        M = cv2.getPerspectiveTransform(pts1,pts2)        
        dst = cv2.warpPerspective(img,M,(300,300))
        cv2.imshow('after',dst)
        
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

points=[]

def draw_circle(event,x,y,flags,param):
    global points
    
    if event==cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(img,(x,y),10,(255,255,255),-1)
        print(x,y)
        points.append([x,y])
        
img=cv2.imread('arashi.png')
cv2.namedWindow('arashi')
cv2.setMouseCallback('arashi',draw_circle)

while(1):
    cv2.imshow('arashi',img)
    if cv2.waitKey(20)%0xFF==27:
        break
    if len(points)==4:
        pts1=np.float32(points)
        pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
        M=cv2.getPerspectiveTransform(pts1,pts2)
        dst=cv2.warpPerspective(img,M,(300,300))
        cv2.imshow('result',dst)
        
cv2.destroyAllWindows()
```

```python
#Morphological Transform #erosion
img = cv2.imread('./images/j.png',0)

kernel = np.ones((5,5),np.uint8)

erosion=cv2.erode(img,kernel,iterations=1)

plt.figure(figsize=(12,12))
plt.subplot(121); plt.imshow(img)
plt.subplot(122); plt.imshow(erosion)
```

```python
#dilation
img = cv2.imread('./images/j.png',0)

kernel = np.ones((5,5),np.uint8)

dilation=cv2.dilate(img,kernel,iterations=3)

plt.figure(figsize=(12,12))
plt.subplot(121); plt.imshow(img)
plt.subplot(122); plt.imshow(dilation)
```

```python
#salt and pepper noise 파악해볼 것
row,col = img.shape
s_vs_p = 1 #0.5
amount = 0.01
out = np.copy(img)
# Salt mode
num_salt = np.ceil(amount * img.size * s_vs_p)
coords = [np.random.randint(0, i, int(num_salt))
          for i in img.shape]
out[tuple(coords)] = np.max(img)
        
# Pepper mode
num_pepper = np.ceil(amount* img.size * (1. - s_vs_p))
coords = [np.random.randint(0, i, int(num_pepper))
          for i in img.shape]
out[tuple(coords)] = np.min(img)

opening=cv2.morphologyEx(out,cv2.MORPH_OPEN,kernel)
plt.figure(figsize=(10,5))
plt.subplot('121')
plt.imshow(out,cmap='gray'), plt.axis('off')
plt.subplot('122')
plt.imshow(opening,cmap='gray'), plt.axis('off')
```

```python
row,col = img.shape
s_vs_p = 0 #~0.5 : pepper // 0.5 ~ 1 : salt noise
amount = 0.01
out = np.copy(img)
# Salt mode
num_salt = np.ceil(amount * img.size * s_vs_p)
coords = [np.random.randint(0, i, int(num_salt))
          for i in img.shape]
out[tuple(coords)] = np.max(img)
        
# Pepper mode
num_pepper = np.ceil(amount* img.size * (1. - s_vs_p))
coords = [np.random.randint(0, i, int(num_pepper))
          for i in img.shape]
out[tuple(coords)] = np.min(img)

opening=cv2.morphologyEx(out,cv2.MORPH_CLOSE,kernel)
plt.figure(figsize=(10,5))
plt.subplot('121')
plt.imshow(out,cmap='gray'), plt.axis('off')
plt.subplot('122')
plt.imshow(opening,cmap='gray'), plt.axis('off')
```

```python
plt.figure(figsize=(12,12))
plt.subplot(121); plt.imshow(img, cmap='gray')

out=img.copy()
gradiant=cv2.morphologyEx(out,cv2.MORPH_GRADIENT,kernel)

plt.subplot(122); plt.imshow(gradiant, cmap='gray')
```
